```python
from pathlib import Path
BASE_DIR = Path.cwd()          # ml/
DATA_DIR = BASE_DIR.parent / "data"

def load_csv(name):
    return pd.read_csv(DATA_DIR / f"{name}.csv")
```

```python
orders=load_csv("order_level")
fact=load_csv("fact")
products=load_csv("dim_products")
```

```python
new = fact.groupby(['order_id','order_item_id']).agg({
    'price': 'sum',                # sum the price
    'seller_id': lambda x: list(x) # collect all sellers in a list
}).reset_index()
```
